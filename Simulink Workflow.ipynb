#### Task 1.

In the first part of this project, you will obtain an LTI system by linearizing the plant: the robotic leg.
 
The model at right represents leg dynamics. Assume the leg rotates from -30 degrees to 30 degrees in a single step. The robotic leg is at -30 degrees at time 0. Zero torque is applied to the hip so the only torque acting on the system is from gravity.

Specify the signal _input_torque_ as the input perturbation analysis point.

#### Task 2.

Specify the signal _theta_deg_ as the output measurement analysis point.

#### Task 3.

Now that you have defined the control input and output, you are ready to linearize. By simulating the model, you can see the plant reaches its steady state, or equilibrium position, after about 3 seconds. In this task, you will linearize the plant about its equilibrium using Model Linearizer.

Use Model Linearizer to obtain an LTI estimation of the plant at operating point t = 4. Then export the LTI system to MATLAB.

#### Task 4.

In this part, you will build a feedback control system with a PID controller and tune the controller against design requirements using PID Tuner. The controller rotates the robotic leg from its steady state (equilibrium) position, 0 degrees, to a small angle, 10 degrees.

Add a PID Controller block (Simulink > Continuous) and connect the block with the _error_ signal and the _torque_ signal. Connect the signal _theta_ to the Signal Assessment block.

#### Task 5.

As you have already learned, you can use the Response Time and Transient Behavior sliders in the PID Tuner app to reach your control goals. The Show Parameters option shows you the values of your tuned system characteristics while you tune.

Open PID Tuner by choosing Tune… in the PID Controller block dialog.

Tune your system to the following requirements:
* Settling time < 0.8 seconds
* Overshoot < 10%

Click Update Block to export the tuned parameters to your model.
